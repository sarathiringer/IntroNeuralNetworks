# 3. Stochastic Gradient Descent

How can we find the minimum in many dimensions in an efficient way? Let's introduce ***stochastic gradient descent***!

Imagine now if I want to go down to the lowest point below as fast as possible (and I'm not concerned about human things like falling and injuring myself). In what direction should I go?

The idea of measuring the direction in which the mountain is the steepest is the very core of the minimization method of *stochastic gradient descent*. You can also think of it as dropping a ball and have it roll down to the lowest point it can find. If you do it many times, over and over, you are likely to find the global minimum of the function. 

gradient descent.png

In the picture above, borrowed from the [Deep Learning course at Coursera](https://www.coursera.org/learn/machine-learning), $\theta_0$ and $\theta_1$ are parameters (in our case weights and biases) and $J(\theta_0, \theta_1)$ is the loss function that we are looking to minimize. For each cross, we are checking in what directions the function would get smaller the fastest, and then we are taking a step in that direction. The size of the step, wether to go really long or really short each time, is called *learning rate*. When there is no direction that would make the function smaller than it already is, we have reached a minimum. If the steps are really small, we do not risk "missing" (jumping over) the minimum, but it is costly to calculate over and over. With really big steps, we are more efficient but we run in to the risk of jumping from side to side in a valley and never reaching the bottom.

**The point of the optimization of the network is to have it learn its parameters so that it best mimics the data - meaning mapping the input to the output. Stochastic gradient descent is a way to do this.**
