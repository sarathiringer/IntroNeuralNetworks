# ReLU - Rectified Linear Unit

Skärmavbild 2021-08-14 kl. 13.51.16.png


```python
# Defining the ReLU activation function
def ReLU(x):
  if x < 0:
    y = 0
  else:
    y = x
  return y
```

```python
# Define a range of input
x = np.arange(-10, 10, 1)
print(f"input: {x}")
```

```python
# Send the inputs through the activation function
y = []
for i in x:
  y.append(ReLU(i))
print(f"output: {y}")
```

```python
# Plot the result
px.line(x=x, y=y, title="ReLU activation function")
```

# Sigmoid function



Skärmavbild 2021-08-14 kl. 14.06.01.png

```python
# Define the sigmoid function
def sigmoid(x):
	y = (1 / (1 + np.e**(-x)))
	return y
```

```python
# Define a range of input
x = np.arange(-10, 10)
print(f"input:", x)
```

```python
# Send the inputs through the activation function
y = []
for i in x:
  y.append(sigmoid(x=i))
print(f"output from the activation function:", y)
```

```python
# Plot the result
px.line(x=x, y=y, title='Sigmoid activation function')
```

# Soft ReLU / softplus / Smooth ReLU

Skärmavbild 2021-08-14 kl. 14.16.55.png

```python
# Define the soft ReLU function
def soft_ReLU(x):
	y = np.log(1 + np.e ** x)
	return y
```

```python
# Define a range of input
x = np.arange(-10, 10)
print(f"input:", x)
```

```python
# Send the inputs through the activation function
y = []
for i in x:
  y.append(soft_ReLU(x=i))
print(f"output from the activation function:", y)
```

```python
# Plot the result
px.line(x=x, y=y, title='Soft ReLU activation function')
```
